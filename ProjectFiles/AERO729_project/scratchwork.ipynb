```python
from particle import Particle
import numpy as np
from simulationenv import SimulationEnviroment
import matplotlib.pyplot as plt
```

Create simulation enviroment

```python
s = SimulationEnviroment(
    particles=100,
    env_height=1,
    env_width=1,
    tolerance=1e-3,
    sigma=0.01,
    epsilon=0.01
)
```

Run gradient descent to find intial configuration

```python
s.steepest_descent_initialization()
```

```python
s.run_sim(
    duration=10,
    del_t=1e-4
)
```
