```python
from particle import Particle
import numpy as np
from simulationenv import SimulationEnviroment
import matplotlib.pyplot as plt
```

```python
s = SimulationEnviroment(
    particles=10,
    env_height=1,
    env_width=1,
    tolerance=1e-3,
    sigma=0.01,
    epsilon=0.01
)
```

```python
s.run_sim(
    duration=1,
    del_t=1e-4
)
```

```python
s1 = s.particle_list[1]
s2 = s.particle_list[2]
```

```python
x1, y1 = zip(*s1.trajectory)
x1 = np.array(x1)
y1 = np.array(y1)
x2, y2 = zip(*s2.trajectory)
x2 = np.array(x2)
y2 = np.array(y2)
```
